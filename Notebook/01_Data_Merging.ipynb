```python
installments["PAYMENT_DIFF"] = installments["AMT_PAYMENT"] - installments["AMT_INSTALMENT"]
installments["PAYMENT_DELAY"] = installments["DAYS_ENTRY_PAYMENT"] - installments["DAYS_INSTALMENT"]
inst_agg = installments.groupby("SK_ID_CURR").agg({
    "AMT_INSTALMENT": ["mean", "sum"],
    "AMT_PAYMENT": ["mean", "sum"],
    "NUM_INSTALMENT_NUMBER": "count",
     "PAYMENT_DIFF": ["mean", "min", "max"],
    "PAYMENT_DELAY": ["mean", "min", "max"]
})

inst_agg.columns=[f"inst_{col[0]}_{col[1]}" for col in inst_agg.columns]
inst_agg.reset_index(inplace=True)
    
```

```python
# Fusion dans application_train
app = app.merge(inst_agg, on="SK_ID_CURR", how="left")
```

```python
pos_agg = POS.groupby("SK_ID_PREV").agg({
    "MONTHS_BALANCE": ["min", "max", "count"],
    "CNT_INSTALMENT": "mean",
    "CNT_INSTALMENT_FUTURE": "mean"
})
pos_agg.columns = [f"pos_{col[0]}_{col[1]}" if col[1] else f"pos_{col[0]}" for col in pos_agg.columns]
pos_agg.reset_index(inplace=True)
previous = previous.merge(pos_agg, on="SK_ID_PREV", how="left")
```

```python
# AgrÃ©gation Credit par SK_ID_PREV
credit_agg = credit.groupby("SK_ID_PREV").agg({
    "MONTHS_BALANCE": ["min", "max", "count"],
    "AMT_BALANCE": "mean",
    "AMT_CREDIT_LIMIT_ACTUAL": "mean",
    "AMT_DRAWINGS_CURRENT": "mean"
})

# Renommage des colonnes
credit_agg.columns = [f"credit_{col[0]}_{col[1]}" if col[1] else f"credit_{col[0]}" for col in credit_agg.columns]
credit_agg.reset_index(inplace=True)

# Fusion directe avec previous_application
previous = previous.merge(credit_agg, on="SK_ID_PREV", how="left")
```

```python
# AgrÃ©gation de previous_application enrichi par SK_ID_CURR
prev_agg = previous.groupby("SK_ID_CURR").agg({
    "AMT_ANNUITY": ["mean", "max"],
    "AMT_APPLICATION": ["mean", "max"],
    "AMT_CREDIT": ["mean", "max"],
    "NAME_CONTRACT_TYPE": lambda x: x.value_counts().index[0],  # le type de contrat le plus frÃ©quent
    "pos_MONTHS_BALANCE_min": "mean",  # Exemples de features POS/Credit que tu as ajoutÃ©es
    "credit_AMT_BALANCE_mean": lambda x: x.mean()
})

# Renommage des colonnes comme tu aimes
prev_agg.columns = [f"prev_{col[0]}_{col[1]}" if col[1] else f"prev_{col[0]}" for col in prev_agg.columns]
prev_agg.reset_index(inplace=True)

# Fusion finale avec application_train
app = app.merge(prev_agg, on="SK_ID_CURR", how="left")
```
